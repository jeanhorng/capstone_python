```python
url = "https://www.notion.so/try-3413f87d9b0f449bb3d2ee8a3b8b80d0"
def check_site():
    N=0

    while N<20:
        rs = requests.get(url)
        now = datetime.now()
        current_time = now.strftime("%H:%M:%S")
        if rs.status_code == 200:

            print(current_time, "OK")
            N+=1
            time.sleep(5)
            
            pass
        else:
            print(current_time,"sth's wrong")
            break
    
```
