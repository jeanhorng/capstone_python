```python
"""
**Patient / Doctor Scheduler** - Create a patient class and a doctor class. 
Have a doctor that can handle multiple patients and setup a scheduling program where a doctor 
can only handle 16 patients during an 8 hr work day.
"""
```

```python
from IPython.display import clear_output
```

```python
#patient: choose doc,time.
#doctor: doc_schedule: 2 patients/1hr of 8 hrs
#scheduling: patient choose a doc, check doc schedule (16patients a day / 8hrs),
class Patient:
    def __init__(self,name):
        self.name=name
    def setappointment(self,schedule):
        self.schedule = schedule
        
    def show_myappoint(self):
        return self.schedule

    
        
        

class Doctor:
    def __init__(self, name):
        self.name = name
        self.schedule_by_hour=dict()
        for n in range(8,17):
            if n != 12:
                self.schedule_by_hour[n] = 0
        self.today_patients = []
    def hourly_schedule(self):

        return self.schedule_by_hour
    def set_schedule(self,p_name,slot):
        self.schedule_by_hour[slot] += 1
        self.today_patients.append([p_name,slot])
    def todays_patient(self):
        return self.today_patients


    
        
    

    
    
```

```python
def p_scheduling(my_doc):
    doc_schedule = my_doc.hourly_schedule()
    
    while True:
        clear_output()
        p_name = input('Enter your name plz: ')
        p_name = p_name.capitalize()
        patient = Patient(p_name)
        
        scheduling = True
        while scheduling:
            clear_output()
            print('\n',pick_doc +"'s schedule for today'" '\n')         
            print(doc_schedule)
            print('(hour:patient(s)). Max: 2 patients an hour')
            try:
                pick = input('Which time slot do you want?')
                pick = int(pick)
                if doc_schedule[pick] <2:
                    my_doc.set_schedule(p_name, pick)               

                    p_schedule = 'Patient: '+ p_name + ', Time: ' + str(pick) + "o'clock " + ', Doctor: '+ my_doc.name 
                    print('\nScheduled!')
                    patient.setappointment(p_schedule)
                    break

                else:
 
                    continue
            except:

                continue

        break
    patients_doc[p_name] = [patient, my_doc]


```

```python
def check_info():
    while True:

        ur_name = input("Enter your name to check info: ")
        ur_name = ur_name.capitalize()
        if ur_name in patients_doc.keys():
            patient = patients_doc[ur_name][0]
            print('\n',patient.show_myappoint())
            doc = patients_doc[ur_name][1]
            print("Doc " + doc.name + "'s schedule for today"+": \n" + str(doc.hourly_schedule()))
        else:
            continue
            
        go_on = input('Continue checking?')
        if go_on.lower() == 'y':
            continue
        else:

            print('See you next time.')
            break
```

```python

#create doctors
Jean = Doctor('Jean')
Jane = Doctor('Jane')
Ann = Doctor('Ann')

doctors = {'Jean':Jean,'Jane':Jane,'Ann':Ann}
doc_names =[]
for name, doc in doctors.items():
    doc_names.append(name)

patients_doc = dict()

while True:
    clear_output()

    #select doc and see if there's such doc
    pick_doc = input('which doc do you want to visit:' + str(doc_names))
    pick_doc = pick_doc.capitalize()
    if pick_doc in doc_names:
        my_doc = doctors[pick_doc]
        pass

    else:
        continue
        
    #patient enter info and choose time
    p_scheduling(my_doc)
    
    go_on = input('Continue scheduling? Y or N')
    if go_on.lower() == 'y':
        
        continue
    else:
        break

check_info()
        




            
    
```
