```python
"""
**Product Inventory Project** - Create an application which manages an inventory of products. 
Create a product class which has a price, id, and quantity on hand. Then create an *inventory* class 
which keeps track of various products and can sum up the inventory value. [[Drhealsgood (Python)]]
(https://github.com/Drhealsgood/miniprojects/blob/master/class_projects/product_inventory/product_inventory.py) 
"""
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
class product:
    def __init__(self, name, id_num, price, quantity_on_hand):
        self.name = name
        self.id_num = id_num
        self.price = price
        self.quantity_on_hand = quantity_on_hand
```

```python


class inventory:
    def __init__(self):
        self.product_list = []
    def new_item(self,*product):
        list(product)
        self.product_list.extend(product)  
    def product_info(self):
        items = []
        id_num = []
        quantity = []
        item_price = []
        value = []
        item_info =dict()

        for p in self.product_list:

            items.append(p.name)
            id_num.append(p.id_num)
            quantity.append(p.quantity_on_hand)
            item_price.append(p.price)
            value.append(p.quantity_on_hand*p.price)
            
        item_info['Name'] = items
        item_info['ID'] = id_num
        item_info['Quantity'] = quantity
        item_info['price'] = item_price
        item_info['Value'] = value

        df = pd.DataFrame(item_info,index=range(1,len(items)+1))
        df.loc["Total quantity & value"] = pd.Series(data=[df['Quantity'].sum(),df["Value"].sum()],index=["Quantity","Value"])
        print(df)


```
