```python

def FF(n):
    a = 0
    b = 1
    c = 1
    if n == 0:
        return a
    elif n == 1 or n ==2:
        return b
    elif n>2:
        for i in range(n-2):

            d = b + c 
            #e = c+d
            #f = d+e
            b = c
            c = d
    return d
    
    
```
