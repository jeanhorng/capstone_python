```python
"""
**Page Scraper** - Create an application which connects to a site and pulls out all links, or images, and saves 
them to a list. *Optional: Organize the indexed content and donâ€™t allow duplicates.
Have it put the results into an easily searchable index file.*
"""
```

```python
from bs4 import BeautifulSoup, SoupStrainer
import requests
import pandas as pd
```

```python
url ="https://www.theguardian.com/international"
rs = requests.get(url)
soup = BeautifulSoup(rs.text,'lxml')
titles=[]
lks=[]
links = soup.select('.fc-item__container a')

for link in links:
    titles.append(link.text)
    lks.append(link['href'])
    
d = dict()
d['Titles'] = titles
d['Links'] = lks

news = pd.DataFrame(data=d)
news = news.drop_duplicates(subset=['Links'])
```

```python
def findnews(title):
    print(news[news['Titles'].str.contains(title)])
```
