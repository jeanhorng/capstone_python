```python
from bs4 import BeautifulSoup, SoupStrainer
import requests
import pandas as pd
from IPython.display import HTML
```

```python
url ="https://www.theguardian.com/international"
rs = requests.get(url)
soup = BeautifulSoup(rs.text,'lxml')
titles=[]
lks=[]
links = soup.select('.fc-item__container a')

for link in links:
    titles.append(link.text.lower())
    lks.append(link['href'])
    
d = dict()
d['Titles'] = titles
d['Links'] = lks

news = pd.DataFrame(data=d)
news = news.drop_duplicates(subset=['Links'])

# link is the column with hyperlinks
def make_clickable(val):
    # target _blank to open new window
    return '<a target="_blank" href="{}">{}</a>'.format(val, val)

#news = news.style.format({'Links': make_clickable})
news['Links']=news['Links'].apply(make_clickable)
N = HTML(news.to_html(escape=False))
print(type(news))
print(type(N))
```

```python

def findnews():
    pd.options.display.max_colwidth = 1000
    pd.set_option('display.max_rows', None)
    title = input('Enter a key word to search or enter "A" to display all titles: ')
    if title.lower()== 'a':
        return N
    else:
        if news[news['Titles'].str.contains(title.lower())].size==0: 
            print('Not found')
        else:
            news_selected = news[news['Titles'].str.contains(title.lower())]
            return HTML(news_selected.to_html(escape=False))
        
            
```
