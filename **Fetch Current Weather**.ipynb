```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
import time
```

```python
def get_location():
    rs = requests.get("http://ip-api.com/json/")
    result = rs.json()
    infos = [ 'country', 'city']
    infols =[]
    for info in infos:
        infols.append(result.get(info))
    city_country = infols[1] + " " + infols[0]
    return city_country
    
```

```python
def check_current_weather():
    url ='https://www.google.com.tw/'
    options = Options()
    options.headless = True

    chrome = webdriver.Chrome(executable_path='/Users/jean/Downloads/chromedriver', options=options)
    chrome.get(url)
    elem = chrome.find_element_by_css_selector('input')
    
    location = get_location()
    
    elem.send_keys('current weather'+ location , Keys.ENTER)
    chrome.implicitly_wait(10)
    elem = chrome.find_element_by_class_name('wob_t.TVtOme')
    degree_ele = chrome.find_element_by_xpath('//span[@aria-label="°Celsius"]')

    return location + ": "+ elem.text + degree_ele.text
```
