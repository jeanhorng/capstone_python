```python
"""
**Fizz Buzz** - Write a program that prints the numbers from 1 to 100. But for multiples of three print “Fizz” 
instead of the number and for the multiples of five print “Buzz”. 
For numbers which are multiples of both three and five print “FizzBuzz”.
"""
```

```python
def fizzbuzz():
    num_list = []
    for n in range(1,51):
        if n%3 == 0 and n%5 == 0:
            n = 'FizzBuzz'
        elif n%3 == 0:
            n = 'Fizz'
        elif n%5 == 0:
            n = 'Buzz'
        num_list.append(n)
    print(num_list)
            
    
            
            
```

```python
fizzbuzz()
```

```python
"""
**Reverse a String** - Enter a string and the program will reverse it and print it out. [[Drhealsgood (Python)]]
"""
```

```python
def reverse():
    string = input("Enter a string and I will reverse it: ")
    n = (len(string))
    reversed_str = ''
    for i in range(1,n+1):
        reversed_str += string[n-i]
    return reversed_str
    
```

```python
reverse()
```

```python
def reverse2():
    string = input("Enter a string and I will reverse it: ")
    return string[::-1]
```

```python
def check():
    word = input('Enter a word and check if it is a palindrome: ')
    if word[::] == word[::-1]:
        print('Yes, it is.')
    else:
        print('No, it isnt.')
```

```python
check()
```

```python
def palindrome(string):
	num = len(string) - 1
	second_half = ''
	mid = len(string) - 1
	while (num > mid/2):
		second_half += string[num]
		num -=1

	first_half = string[:len(string)/2]
	if first_half == second_half:
		return True
	else:
		return False

```

```python
palindrome("racecar")
```

```python

def count_words(textfile):
    str1=''
    str2=''
    str3=string.punctuation+'\n'
    opentext = open(textfile, 'r+')
    text = opentext.read()
    table=text.maketrans(str1,str2,str3)
    print('tablbe',table,'\n')
    no_punc_text = text.lower().translate(table)
    print('no', no_punc_text.split(),'\n', len(no_punc_text.split()))
    unique_text = list(set(no_punc_text.split(' ')))
    print(unique_text)
    opentext.close()
    return 'The num of unique words are: ' + str(len(unique_text)) 
```

```python
count_words("text_file.txt")
```

```python
def count_words(text):
	return str(len(text.split(" "))) + " words in the analyzed text"

def count_from_file(file_location):
	file = open(file_location, 'r')
	text = str(file.read())
	return count_words(text)
```
