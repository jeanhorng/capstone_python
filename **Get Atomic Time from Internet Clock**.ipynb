```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
```

```python
options = Options()
options.headless = True
chrome = webdriver.Chrome(executable_path = '/Users/jean/Downloads/chromedriver', options=options)
chrome.get('https://www.time.gov/')
eles = chrome.find_element_by_css_selector('#main-time-area')
print(eles.text)
```
