```python
import getpass
```

```python

#create a checking account with initial deposit of $400
#initial deposit, set password, do actions(deposit,withdraw,display)

class Account:
    def __init__(self, cust_id, initial_deposit):
        self.id = cust_id
        self.password = getpass.getpass("Hi "+self.id.upper() + ", Please create your password: ")
        id_ps = {self.id:self.password}
        self.initial_deposit = initial_deposit
        self.balance = self.initial_deposit
        print("An account of", self.id,'is created.')
        print('The existing balance is:', self.balance)

            

class CheckingAccount(Account):
    def deposit(self, deposit_money):

        self.deposit_money = deposit_money
        self.balance += self.deposit_money
        print('deposit successful')
        print('\ncurrent balance is:', self.balance)
        
    def withdrawal(self, withdrawal_money):

        if withdrawal_money <= self.balance:
            self.withdrawal_money = withdrawal_money
            self.balance -= self.withdrawal_money
            print('\ncurrent balance is:', self.balance)
        else:
            print('\nNot enough balance!')
            
    def display_balance(self):
        print("\ncurrent balance:", self.balance)


def initialize_objects():
    account = input('Creating checking account. Plz enter your name: ')
    account_holder = account.lower()
    checking_account = CheckingAccount(account_holder,400)
    #sally_checking = CheckingAccount('sally',400)
    #billy_checking = CheckingAccount('billy',300)
    
    master_dict = {account_holder:checking_account}
    
    option = input('Do you want to use ATM and insert ur card? Y or N')
    
    if option.lower() == 'y':
        #check isCustomer
        while True:
            name = input('Card inserted. Enter your name plz: ')
            if name.lower() in master_dict:
                cust_account = master_dict[name.lower()]
                break
            else:
                print('Name not recognized')
                pass
            
        #check password
        n=0
        while n<=3:
            password = getpass.getpass('Enter password please: ')
            if password == cust_account.password:
                break
            else:
                print('Wrong ps. Try again. You have',3-n, 'trials left')
                n += 1
        if n>3:
            return "The card is locked. Please contact your bank."
        
        #choose action
        else:
            
            while True:
                option = input('Press 1 for deposit, 2 for withdrawal, 3 for display balance, 4 for exit: ')
                if option == '1':
                    make_deposit = input("Enter the amount for deposit: ")
                    if make_deposit.isnumeric() == True:
                        pass
                    else:
                        continue
                    cust_account.deposit(int(make_deposit))
                elif option == '2':
                    make_withdrawal = input("Enter the amount to withdrawal: ")
                    if make_withdrawal.isnumeric() == True:
                        pass
                    else:
                        print('Is not numeric')
                        continue
                    cust_account.withdrawal(int(make_withdrawal))
                elif option == '3':
                    cust_account.display_balance()
                elif option == '4':
                    break
                else:
                    print('\nPlease press 1,2 3, or 4')
                    continue
                
                option = input('Do you want to do anything else? Y or N: ')
                if option.lower() == 'n':
                    break
                else:
                    continue                                                            
    else:
        pass
    
    return 'See you next time.'
                
            

            

        
```
