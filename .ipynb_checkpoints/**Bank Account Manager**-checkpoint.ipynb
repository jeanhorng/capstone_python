```python
class Animal:
    def __init__(self,name):
        self.name = name
        print("An instance of Animal is created")
    def speak(self):              # Abstract method, defined by convention only
        raise NotImplementedError("Subclass must implement abstract method")
    def walk(self):
        raise NotImplementedError("Subclass must implement abstract method")
    
class Dog(Animal):
    def run(self):
        return "run"
```

```python
ff = Dog('Henry')
```

```python
ff.name
```

```python
ff.walk()
```

```python
ff.speak()
```
